How the quality of methods reporting in experiments is related to citation of the publication?

- 23 papers were identified and included. 
- We defined a checklist of essential parameters that should be reported and have scored the number of those parameters that are reported for each publication. 
- Bibliometric parameters (impact factor, citations and h-index) were used to look for association between Journal and Author status and the quality of method reporting.
