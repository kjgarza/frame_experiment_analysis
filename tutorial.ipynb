How the quality of methods reporting in experiments is related to publication citations?

- 23 papers were identified and included.

#### Loading some libraries 

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from pylab import *
```

#### Loading the Data

```python
input_data = pd.DataFrame.from_csv('quality_methods_reporting.csv')
```

```python
input_data
```

#### Correlation Analysis

> The Pearson product-moment correlation coefficient is a measure of the strength of the linear relationship between two variables. It is referred to as Pearson's correlation or simply as the correlation coefficient.

```python
stats.pearsonr(input_data['Evaluator 1 (%)'], input_data['Evaluator 2 (%)'])
```

#### Ploting 

```python
plt.scatter(input_data['Evaluator 1 (%)'], input_data['Evaluator 2 (%)'])
```

```python
Image(filename='pone.0101131.g007.jpg')
```

```python
r_2 = stats.pearsonr(input_data['Evaluator 1 (%)'], input_data['Evaluator 2 (%)'])

fit = polyfit(input_data['Evaluator 1 (%)'], input_data['Evaluator 2 (%)'], 1) 
fit_fn = poly1d(fit)

plt.plot(input_data['Evaluator 1 (%)'], input_data['Evaluator 2 (%)'], 'bs', input_data['Evaluator 1 (%)'], fit_fn(input_data['Evaluator 1 (%)']), '--k')

plt.text(60, 90, r'$r^2= $')
plt.text(67, 90, round(r_2[0],2))

plt.ylim([0,100])
plt.xlim([0,100])
```
